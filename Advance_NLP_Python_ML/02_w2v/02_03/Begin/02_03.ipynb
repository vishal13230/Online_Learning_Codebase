```python
message = pd.read_csv("C:/Users/visha/Downloads/Advance_NLP_Python_ML/data/spam.csv", encoding="latin-1")
message.head()
```

```python
message = message.drop(labels=["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"], axis=1)
message.columns = ["label", "text"]
message.head()
```

```python
message["text_clean"] = message["text"].apply(lambda x : gensim.utils.simple_preprocess(x))
message.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(message["text_clean"], message["label"], test_size=0.2)
```

```python
w2v_model = gensim.models.Word2Vec(X_train, vector_size=100, window=5, min_count=2)
```
